# Домашнее задание к лекции "Функции"

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

перечень всех документов

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)

```python
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

Общие требования к программе:

код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
в коде отсутствуют глобальные переменные (за исключением documents и directories);
пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу (вводом команды "q").

Задание 1
Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру
Примеры работы:

1.

Введите команду:
p

Результат:

```python
def p(t):
    """
    Задание 1 Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру
    """
    for i in documents:
        if t == i['number']:
            return i['name']            
```

Пункт 2. Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
Примеры работы:

1.

Введите команду:
s

Введите номер документа:
10006
Результат:
Документ хранится на полке: 2

2.

Введите команду:
p

Введите номер документа:
12345
Результат:
Документ не найден в базе

```python
def s(t):
    """
    Пункт 2. Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
    """
       
    for k, v in directories.items():
        if t in v:
            return k
```

Пункт 3. Пользователь по команде "l" может увидеть полную информацию по всем документам
Пример работы:

Введите команду:
l
Результат:

№: 2207 876234, тип: passport, владелец: Василий Гупкин, полка хранения: 1
№: 11-2, тип: invoice, владелец: Геннадий Покемонов, полка хранения: 1
№: 10006, тип: insurance, владелец: Аристарх Павлов, полка хранения: 2

```python
def l():
    """
    Пользователь по команде "l" может увидеть полную информацию по всем документам
    """
    
    for i in documents:
        print(f"№: {i['number']}, тип: {i['type']}, владелец: {i['name']}, полка хранения: {s(i['number'])}")
```

Пункт 4. Пользователь по команде "ads" может добавить новую полку
Примеры работы:

1.

Введите команду:
ads

Введите номер полки:
10
Результат:
Полка добавлена. Текущий перечень полок: 1, 2, 3, 10.

2.

Введите команду:
ads

Введите номер полки:
1
Результат:
Такая полка уже существует. Текущий перечень полок: 1, 2, 3.


```python
def ads():
    """
    Пользователь по команде "ads" может добавить новую полку
    """
    t = input('Введите номер полки: ')
    
    if t in directories.keys():
        print('Такая полка уже существует.')
    else:
        directories[t] = []
        print('Полка добавлена.')
    print(f"Текущий перечень полок: {', '.join(list(directories.keys()))}")
```

Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
Примеры работы:

1.

Введите команду:
ds

Введите номер полки:
3
Результат:
Полка удалена. Текущий перечень полок: 1, 2.

2.

Введите команду:
ds

Введите номер полки:
1
Результат:
На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: 1, 2, 3.

3.

Введите команду:
ds

Введите номер полки:
4
Результат:
Такой полки не существует. Текущий перечень полок: 1, 2, 3.

```python
def ds():
    """
    Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая) 
    """
    t = input('Введите номер полки: ')
    
    if t not in directories.keys():
        print('Такой полки не существует.')
    elif directories[t] != []:
        print('На полке есть документа, удалите их перед удалением полки.')
    else:
        del(directories[t])
        print('Полка удалена.')
    print(f"Текущий перечень полок: {', '.join(list(directories.keys()))}")
```

Пункт 2. Пользователь по команде "d" может удалить документ из данных

d

d

```python
def d():
    """
    Пользователь по команде "d" может удалить документ из данных
    """
    t = input('Введите номер документа: ')
    check = 0
    
    for n, i in enumerate(documents):
        if t == i['number']:
            del(documents[n])
            check = 1
            print('Документ удален. Текущий список документов: ')
    if check == 0:
        print('Документ не найден в базе. Текущий список документов: ')
    
    l()
```

Пункт 3. Пользователь по команде "m" может переместить документ с полки на полку
Примеры работы:

№: 11-2, тип: invoice, владелец: Геннадий Покемонов, полка хранения: 1
№: 10006, тип: insurance, владелец: Аристарх Павлов, полка хранения: 2

```python
def m():
    """
    Пользователь по команде "m" может переместить документ с полки на полку
    """
    num = input('Введите номер документа: ')
    dr = input('Введите номер полки: ')
    
    if dr not in directories.keys():
        print('Такой полки не существует. Текущий перечень полок: ')
    elif p(num) == None:
        print('Документ не найден в базе. Текущий список документов:')
    else:
        c = s(num)
        tmp = directories[c]
        tmp.remove(num)
        
        directories[c] = tmp
        tmp = directories[dr]
        tmp.append(num)
        directories[dr] = tmp
        
        print('Документ перемещен. Текущий список документов: ')

    l()
```

```python
def main():
    """
    Пользователь по команде "p" может узнать владельца документа по его номеру
    Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
    Пользователь по команде "l" может увидеть полную информацию по всем документам
    Пользователь по команде "ads" может добавить новую полку
    Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
    Пользователь по команде "ad" может добавить новый документ в данные
    Пользователь по команде "d" может удалить документ из данных
    Пользователь по команде "m" может переместить документ с полки на полку
    """
    inp = ''
    while inp != 'q':
        inp = input('Введите команду (для выхода q): ')
        if inp == 'p':
            t = input('Введите номер документа: ')
            if p(t) == None:
                print('Документ не найден в базе')
            else:
                print(f"Владелец документа: {p(t)}")
            
        elif inp == 's':
            t = input('Введите номер документа: ')
            if s(t) == None:
                print('Документ не найден в базе')
            else:
                print(f"Документ хранится на полке: {s(t)}")
        elif inp == 'l':
            l()
        elif inp == 'ads':
            ads()
        elif inp == 'ds':
            ds()
        elif inp == 'ad':
            ad()
        elif inp == 'd':
            d()
        elif inp == 'm':
            m()
        else:
            print('Введена некорректная команда.')
main()
```
