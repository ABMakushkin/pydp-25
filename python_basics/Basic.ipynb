Задание 1
Даны 2 переменных, в которых хранятся строки произвольной длины: phrase_1 и phrase_2.
Напишите код, который проверяет какая из этих строк длиннее.

```python
phrase_1 = input('Введите фразу 1 ')
phrase_2 = input('Введите фразу 2 ')

if len(phrase_1) == len(phrase_2):
    print('Фразы одинаковой длины. Длина каждой фразы составляет', len(phrase_1), 'символов')
elif len(phrase_1) > len(phrase_2):
    print('Фраза 1 длинее фразы 2. Длина фразы 1 составляет', len(phrase_1), 'символов')
else:
    print('Фраза 2 длинее фразы 1. Длина фразы 2 составляет', len(phrase_2), 'символов')
```

Задание 2
Дана переменная, в которой хранится число (год). Необходимо написать программу, которая выведет, является ли данный год високосным или обычным.

Високосный год - год который делится на 4 без остатка
При этом если год делится без остатка на 100 и не делится без остатка на 400, то год не високосный
Если год делится без остатка на 400, то год является високосным.

```python
year = int(input('Введите год '))

if year < 0:
    print('Введено отрицательное значение, попробуйте снова')
elif year == 0:
    print('Начало летоисчиления, введите другой год')
elif year%4 == 0:
    if year%100 == 0 and year%400 != 0:
        print('Обычный год')
    else:
        print('Високосный год')
else:   
   print('Обычный год')
```

Вам нужно написать программу для подбора упаковок по размерам товара. Размеры (ширина, длина, высота) хранятся в переменных 
(в сантиметрах):

Используйте следующие правила:

если каждое из трех измерений менее или равно 15 сантиметрам, то выведите на экран "Коробка №1";
если хотя бы одно из измерений больше 15 сантиметров, но менее 50 сантиметров, то выводите "Коробка №2";
если длина товара больше 2 метров, то выводите "Упаковка для лыж";
во всех остальных случаях выводите "Стандартная коробка №3".

```python
width = int(input('Введите ширину '))
length = int(input('Введите длину '))
height = int(input('Введите высоту '))

if width > 0 and length > 0 and height > 0:
    if width <= 15 or length <= 15 or height <= 15:
        print('Коробка №1')
    elif (width > 15 and width < 50) or (length > 15 and length < 50) or (height > 15 and height < 50):
        print('Коробка №2')
    elif length > 200:
        print('Упаковка для лыж')
    else:
        print('Стандартная коробка №3')
else:
    print('Введено некорректное значение')
```

Задание 5 (необязательное)
Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет "счастливым". Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.

```python
number = input('Введите номер билета ') #т.к. номер может начинаться с 0, то вводим строку, а не число

if len(number) != 6:
    print('Номер билета введен некорректно')
else:
    left_num = int(number[:1]) + int(number[1:2]) + int(number[2:3])
    right_num = int(number[3:4]) + int(number[4:5]) + int(number[5:6])
    if left_num == right_num:
        print('Счастливый билет')
    else:
        print('Несчастливый билет')
```

Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник). Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

если пользователь выбрал круг, запрашиваем его радиус,
если треугольник – длины трех его сторон;
если прямоугольник – длины двух его сторон.
Пример работы программы:

Введите тип фигуры:
Круг

Введите радиус круга:
10
Результат:
Площадь круга: 314.16

Введите тип фигуры:
Треугольник

Введите длину стороны A:
2

Введите длину стороны B:
2

Введите длину стороны C:
3
Результат:
Площадь треугольника: 1.98

```python
import math

type = input('Введите тип фигуры ')
if type.lower() == "круг":
    r = float(input('Введите радиус круга '))
    print('Площадь круга = ', round((math.pi*(r**2)), 2))
elif type.lower() == "прямоугольник":
    a = float(input('Введите длину прямоугольника '))
    b = float(input('Введите ширину прямоугольника '))
    print('Площадь прямоугольника = ', round((a * b), 2))
elif type.lower() == "треугольник":
    a = float(input('Введите длину стороны А '))
    b = float(input('Введите длину стороны В '))
    c = float(input('Введите длину стороны С '))
    p = (a + b + c) / 2
    s = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print('Площадь треугольника = ', round(s, 2))
else:
    print('Некорректно введен тип фигуры. Попробуйте Круг / Прямоугольник / Треугольник')
```
