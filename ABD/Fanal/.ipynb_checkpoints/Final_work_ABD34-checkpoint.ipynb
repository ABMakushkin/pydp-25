```python
import numpy as np
import pandas as pd
from datetime import datetime as dt
# from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
# from sklearn.model_selection import train_test_split
# from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

### Задание 1. Загрузите в колаб файлы по оценкам (ratings) и фильмам (movies) и создайте на их основе pandas-датафреймы

```python
# Загружаем файл с оценками
rating = pd.read_csv('u.data.csv', sep='\t')
rating
```

```python
# загружаем файл с фильмами
movies = pd.read_csv('u.item.csv', sep='|', encoding='ANSI')
movies
```

```python
# в датафрейме есть фильмы с неизвестной категорией
movies[movies['unknown'] == 1]
```

```python
# за данные фильмы было 10 оценок
rating[rating['item id'].isin([267, 1373])]
```

```python
# принимаем, что данными фильмами можно пренебречь для данного исследования, поскольку количество оценок составляет всего 0,1% от общей выбоки
movies = movies[movies['unknown'] != 1]
movies
```

```python
rating = rating[~rating['item id'].isin([267, 1373])]
rating
```

```python
# преобразуем дату релиза из текста в дату и выделяем год релиза
movies['releasedate'] = pd.to_datetime(movies['releasedate'])
movies['release_year'] = movies['releasedate'].dt.year
movies
```

### Задание 2. Средствами Pandas, используя dataframe ratings, найдите id пользователя, поставившего больше всего оценок

```python
# пользователь, поставивший больше всего оценок
top_user_id = rating.groupby(['user id']).agg({'rating': 'count'}).sort_values(by=['rating'], ascending=False).head(1).reset_index()['user id'][0]
top_user_id
```

### Задание 3. Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь 

```python
# фильмы которые оценил данный пользователь
rating_top_user = rating[rating['user id'] == top_user_id]
rating_top_user
```

### Задание 4 Добавьте к датафрейму из задания 3 столбцы: 

    По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет 

    столбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей 

```python
# делаем группировку, чтобы найти общий рейтинг фильмов и общее количество оценок
rating_group = rating.groupby(['item id']).agg({'rating': 'sum', 'user id': 'count' }).reset_index().rename(columns={'rating': 'sum_rating', 'user id': 'rat_count'})
rating_group
```

```python
# датафрейм с жанрами и годом релиза
movies_genge = movies[['movieid', 'Action', 'Adventure', 'Animation', "Children's", 'Comedy',
       'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror',
       'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War',
       'Western', 'release_year']]
```

```python
# джойним датафреймы, чтобы получить итоговый
rating_top_user = rating_top_user.merge(rating_group, how='left', on='item id').\
    merge(movies_genge, how='left', left_on='item id', right_on='movieid').drop(columns=['movieid'])
rating_top_user
```

### Задание 5 Сформируйте X_train, X_test, y_train, y_test 

Для построения модели нам нужны признаки. В качестве таковых будем использовать: 

    Год выхода 
    Жанры 
    Общее количество оценок 
    Суммарную оценку 
    
Предсказываем оценку пользователя

```python
X = rating_top_user[['sum_rating', 'rat_count',
       'Action', 'Adventure', 'Animation', "Children's", 'Comedy', 'Crime',
       'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical',
       'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western',
       'release_year']]
y = rating_top_user['rating']
```

```python
# разделяем выборку на обущающую и тестовую
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=888)
```

### Задание 6. Возьмите модель линейной регрессии (или любую другую для задачи регрессии)  и обучите ее на фильмах 

```python
# модель линейной регрессии
model = LinearRegression()
```

```python
# обучаем модель на тренеровочной выборке
model.fit(X_train, y_train)
```

```python
# делаем предсказания на тестовых данных
y_pred = model.predict(X_test)
y_pred
```

### Задание 7. Оцените качество модели на X_test, y_test при помощи метрик для задачи регрессии 


```python
# Коэффициент детерминации

model.score(X_test, y_test)
```

Коэффициент детерминации получился низким. Модель линейной регрессии не качественная
