Вам необходимо написать функцию, которая будет основана на поиске по сайту habr.com. Функция в качестве параметра должна принимать список запросов для поиска (например, ['python', 'анализ данных']) и на основе материалов, попавших в результаты поиска по каждому запросу, возвращать датафрейм вида:

<дата> - <заголовок> - <ссылка на материал>
В рамках задания предполагается работа только с одной (первой) страницей результатов поисковой выдачи для каждого запроса. Материалы в датафрейме не должны дублироваться, если они попадали в результаты поиска для нескольких запросов из списка.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
request = ['python', 'анализ данных']
request
```

```python
def get_harb_posts():
    
    harb_posts = pd.DataFrame()
    
    for rqst in request:
        params = {
            'q': rqst,
            'target_type': 'posts',
            'order': 'relevance'            
        }
        url = 'https://habr.com/ru/search/'
        req = requests.get(url, params=params).text
        soup = BeautifulSoup(req)
        posts = soup.find_all('div', class_='tm-article-snippet')
                
        for k, el in enumerate(posts):
            date = el.find('time').get('title')[:10]
            title = el.find('h2', class_='tm-article-snippet__title tm-article-snippet__title_h2').text.strip()
            link = 'https://habr.com' + el.find('h2', class_='tm-article-snippet__title tm-article-snippet__title_h2').find('a').get('href')
            title = el.find('h2', class_='tm-article-snippet__title tm-article-snippet__title_h2').text
            
            # проверяем на дубликат новости по заголовку
            if k > 0:
                if len(harb_posts[ harb_posts['title'] == title ]) == 0:
                    pass
                else:
                    continue

            row = {'date': date, 'title': title, 'link': link}
            harb_posts = pd.concat([harb_posts, pd.DataFrame([row])])
            
    return harb_posts.reset_index(drop=True) 
```

```python
get_harb_posts()
```
